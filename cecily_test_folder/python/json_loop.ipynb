make a python dict

```python
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
json_yellow = []
data_type = 'yellow'

for year in range(2009,2025,1):
    for month in months:
        month_dict = {}
        relative_url = f"trip-data/{data_type}_tripdata_{year}-{month}.parquet"
        month_dict['source_relative_url'] = relative_url
        destination_file_name = f'{data_type}_{year}_{month}.parquet'
        month_dict['filename'] = destination_file_name
        json_yellow.append(month_dict)

json_yellow
```

```python
# convert to JSON file

import json

# convert dictionary to JSON string
json_yellow = json.dumps(json_yellow)

# write JSON string to a file
with open('yellow_loop.json', 'w') as json_file:
    json_file.write(json_yellow)
```
